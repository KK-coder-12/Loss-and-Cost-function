```python
import numpy as np
y_predicted=np.array([1,1,0,0,1])
y_true= np.array([0.30,0.7,1,0,0.5])
```

```python
def mae(y_true, y_predicted):
    total_error=0
    for yt,yp in zip(y_true, y_predicted):
        total_error+= abs(yt - yp)
    print("Total Error:",total_error)
    mae= total_error/len(y_true)
    print("MAE:",mae)
    return mae
```

```python
mae(y_true, y_predicted)
```

```python
np.mean(np.abs(y_true - y_predicted))
```

```python
absolute_errors = np.abs(y_true - y_predicted)
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))
plt.plot(range(len(y_true)), absolute_errors, marker='o', linestyle='-', color='crimson', label='Absolute Error')

plt.title("Mean Absolute Error (MAE) per Sample", fontsize=14)
plt.xlabel("Sample Index", fontsize=12)
plt.ylabel("Absolute Error", fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
def mse(y_true, y_predicted):
    total_error=0
    for yt,yp in zip(y_true,y_predicted):
        total_error+=(yt - yp)**2
    print("Total_Error:",total_error)
    mse=total_error/len(y_true)
    print("MSE:",mse)
    return mse
```

```python
mse(y_true, y_predicted)
```

```python
np.mean((y_true - y_predicted)**2)
```

```python
y_pred = np.linspace(0, 2, 100)

# MSE: squared error
mse_loss = (y_true - y_pred) ** 2

# Plotting the U-shaped parabola
plt.figure(figsize=(5, 5))
plt.plot(y_pred, mse_loss, color='blue')

# Add red dotted axis lines like your reference
plt.axvline(x=y_true, color='red', linestyle=':', linewidth=1)  # vertical center line
plt.axhline(y=0, color='red', linestyle=':', linewidth=1)        # horizontal base line

# Title and display
plt.title("Parabola (U-shaped MSE Loss Curve)")
plt.xlabel("Predicted Value")
plt.ylabel("MSE Loss")
plt.grid(True, linestyle='--', alpha=0.4)
plt.show()
```

```python
def log_loss(y_true , y_predicted):
    y_predicted_new=[max(i,eplison) for i in y_predicted]
    y_predicted_new=[min(i, 1-eplison) for i in y_predicted_new]
    y_predicted_new= np.array(y_predicted_new)
    return -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```
